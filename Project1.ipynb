## Question  1: What were the top 5 activities that impacted student to enroll in an computing course at CCM?
- filter for "activity_impact" columns
    - replace "High Impact" and "Some Impact" to numeric for count
- bar gragh to display data

```python
# filter data set for all columns that have 'activity_impact'
impct = []
for col in data.columns:
    if 'activity_impact' in col:
        impct.append(col)
 
# impact is column names
impct

# data frame stored as impact
# check update
impact = data[impct]
impact
```

```python
# change column names of impact, this will make plot more legible later
impact = impact.rename(columns = {'activity_impact_mhs_class': 'Middle/High school class', 'activity_impact_mhs_club': 'Middle/High School Club',
                                   'activity_impact_cmp_comp': 'Computer Competition', 'activity_impact_aft_sch_cmp': 'Afterschool Camp',
                                   'activity_impact_summer_cmp': 'Summer Camp', 'activity_impact_ap_class': 'Ap Class',
                                   'activity_impact_dual_enrlmt': 'Dual Enrollment', 'activity_impact_fmly_frnd': 'Family Friend',
                                   'activity_impact_fmly_cmp': 'Family/Friend in Computing', 'activity_impact_hs_faculty': 'High School Staff',
                                   'activity_impact_employer': 'Employer'})

# check update
impact.columns
```

```python
# convert the entries to 1 if high/some impact or 0 if no impact
# data type is object currently
pd.set_option('future.no_silent_downcasting', True)
update1 = impact.replace('Some Impact', 1)
update2 = update1.replace('No Impact', 0)
impact = update2.replace('High Impact', 1)

# change data type to numeric
# can count columns now 
impact = impact.astype(str).astype(int)

# check update
print(impact.dtypes)
impact
```

```python
# .sum() will sum each column and store their sum
colSum = impact.sum()

# order from 
onesSort = colSum.sort_values()

# check update
onesSort
```

```python
# plotting data, horizontal bar plot
plt.figure(figsize=(10, 6))
plt.barh(onesSort.index, onesSort.values, color = 'salmon')
plt.title('Impact of Events on Computing Course Enrollment', fontsize = 18, loc = 'center')
plt.xlabel('Number of Students')
plt.ylabel('Events')
```

The data above shows that a "Family Friend" was the primary reason that students enrolled in a computing course at CCM. The fifth most popular reason that a student enrolled in a computing course at CCM was from a club in middle or high school. A middle or high school class, an employer, or a friend or family member in computing were the three other main events that encouraged students to enroll.

## Question 3: Over the years, which had the highest percentage of women taking computer courses?
- filter 'gender' column from all years
- calculate number and percentage of women per year
- overlaying plot
    - line plot
    - bar plot

```python
# select gender column from each year and drop any na

# gender 2020
gen20 = df20['Gender']
gen20 = gen20.dropna()

# gender 2021
gen21 = df21['Gender']
gen21 = gen21.dropna()

# gender 2022
gen22 = df22['Gender']
gen22 = gen22.dropna()

# gender 2023
gen23 = df23['Gender']
gen23 = gen23.dropna()

# gender 2024
gen24 = df24['Gender']
gen24 = gen24.dropna()
```

```python
# calculate percentage of women per year

# count of women in each year 
numWom20 = 0
numWom21 = 0
numWom22 = 0
numWom23 = 0
numWom24 = 0

# num of women 2020
for wom in gen20:
    if 'Woman' in wom:
        numWom20 += 1

# num of women 2021
for wom in gen21:
    if 'Woman' in wom:
        numWom21 += 1

# num of women 2022
for wom in gen22:
    if 'Woman' in wom:
        numWom22 += 1

# num of women 2023
for wom in gen23:
    if 'Woman' in wom:
        numWom23 += 1

# num of women 2024
for wom in gen24:
    if 'Woman' in wom:
        numWom24 += 1

percent20 = round((numWom20 / gen20.size) * 100, 3)
percent21 = round((numWom21 / gen21.size) * 100, 3)
percent22 = round((numWom22 / gen22.size) * 100, 3)
percent23 = round((numWom23 / gen23.size) * 100, 3)
percent24 = round((numWom24 / gen24.size) * 100, 3)
```

```python
# add number of women to data frame
numWomen = pd.DataFrame({'2020': numWom20, '2021': numWom21, '2022': numWom22, '2023': numWom23, '2024': numWom24}, index = [0])
print('Number of Women:\n', numWomen)

# add women percentage to data frame
women = pd.DataFrame({'2020': percent20, '2021': percent21, '2022': percent22, '2023': percent23, '2024': percent24}, index = [0])
print('\nPercent of Women:\n', women)
```

```python
# because there's a difference in the sample size, one plot will show percentage
# another plot will show the number of women that are in the year
plt.figure(figsize=(10, 6))

# .subplots() to allow multiple plots to stack with each other
fig, ax = plt.subplots()

# plot 1: percentage of women
ax.plot(women.columns, women.iloc[0], marker = '8', linestyle = '--', color = 'goldenrod' )
ax.set_title('Women in Computer Courses Over the Years (2020\'-2024\')')
ax.set_xlabel('Year')
ax.set_ylabel('Percent of Women (%)', color = 'darkgoldenrod')
ax.set_yticks(np.arange(45, 80, 5))

# plot 2: number of women
# .twinx() to share x-axis from plot 1
ax1 = ax.twinx()
# alpha to set transparency behind plot 1
ax1.bar(numWomen.columns, numWomen.iloc[0], color = 'slategrey', alpha  = 0.3)
ax1.set_ylabel('Number of Women', color = 'slategrey')
ax1.set_yticks(np.arange(0, 130, 10))
```

The plot above shows two correlating data for the representation of women in computer courses at CCM. 


The bar plot in gray represents the number of women that are in computer courses from the years of 2020-2024. In 2020, the number of women in computer courses was almost twice the average for the subsequent years of 2021-2024. Since 2020, the number of women in computer courses has taken a decreae.



The line plot represents the percentage of women that are taking computer courses in the years of 2020-2024. Although the year 2020 had the most amount of women taking courses, the same year had the third-highest percentage of women in courses. The year 2024 had the second least amount of women taking computer courses, but that year had the highest percent of women taking crses.
